```python
from typing import Dict, TypedDict
from langgraph.graph import StateGraph
```

```python
# create the state

class AgentState(TypedDict):
    name: str  
```

```python
# Create a node 
def compliment_agent(state: AgentState)-> AgentState:
    """ A node that compliments  the input name"""
    state["name"] = f"{state['name']} you're doing an amazing job learning LangGraph"
    return state
```

```python
# create the graph
graph = StateGraph(AgentState)
graph.add_node("complimenter", compliment_agent)
graph.set_entry_point("complimenter")
graph.set_finish_point("complimenter")

app = graph.compile()
```

```python
app.get_graph().draw_mermaid_png(output_file_path= "excercise1.png")
```

```python
result = app.invoke({"name":"sam"})
```

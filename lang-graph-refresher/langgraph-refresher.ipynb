```python
from typing import Dict, TypedDict
from langgraph.graph import StateGraph # framework that helps to manage the flow of taks in your app using a graph
```

```python
# Create an Agent State - state is shared datastructure
class AgentState(TypedDict): # Our State Schema
    message: str 

# Define a node
# the input to a node is state and output is also a state
def greeting_node(state: AgentState) -> AgentState:
    """Simple Node that add greeting message to state"""
    state['message'] = f" hey {state['message']}, how are you?"
    return state
    
```

```python
graph = StateGraph(AgentState)
graph.add_node("greeter", greeting_node)
graph.set_entry_point("greeter")
graph.set_finish_point("greeter")

app = graph.compile()
```

```python
app.get_graph().draw_mermaid_png(output_file_path="graph.png")
```

```python
# to run
result = app.invoke({"message": "bob"})
```

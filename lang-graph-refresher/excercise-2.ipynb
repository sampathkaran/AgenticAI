```python
from typing import Dict, TypedDict
from langgraph.graph import StateGraph # framework that helps to manage the flow of taks in your app using a graph
```

```python
# define the schema for state 
class AgentState(TypedDict):
    name: str
    values: list[int]
    operation: str
    result: str
    
```

```python
# Create the node
def operation_node(state: AgentState) -> AgentState:
    """ This node will perform math operations"""
    if state["operation"] == "+":
        state['result'] = f"Hi {state['name']}, your answer is {sum(state['values'])}"
    return state
```

```python
# create a node
graph = StateGraph(AgentState)
graph.add_node("operations",operation_node )
graph.set_entry_point("operations")
graph.set_finish_point("operations")

app = graph.compile()
```

```python
output = app.invoke({"name" : "sam", "values" : [1,2,3,4], "operation" : "+" })
```
